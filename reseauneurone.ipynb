```python
import pandas as pd
from sklearn.model_selection import train_test_split

```

```python
df = pd.read_excel('heart_dataset_20.xlsx')
```

```python
print(df.columns)
```

```python
X = pd.get_dummies(df.drop('Heart Disease', axis=1))

y = df['Heart Disease'].apply(lambda x: 1 if x == 'yes' else 0)
```

```python
X = X.astype('float32')
y = y.astype('float32')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
y_train.head()
```

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

```python
model = Sequential()
model.add(Dense(units=32, activation='relu', input_dim=len(X_train.columns)))
model.add(Dense(units=16, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
print(len(X_train), len(X_test))
```

```python
model.fit(X_train, y_train, epochs=200, batch_size=32,validation_split=0.2)
```

```python
y_last = model.predict(X_test)
y_last = (y_last > 0.5).astype(int).flatten()
```

```python
accuracy_score(y_test, y_last)
```
